```python
import pandas as pd
from plotnine import ggplot, aes, geom_line
base = pd.read_csv('teste.csv')
(
    ggplot(base)  # What data to use
    + aes(x="actor", y="movie")  # What variable to use
    + geom_line()  # Geometric object to use for drawing
)
```

```python
import pandas as pd
from plotnine import ggplot, aes, geom_line, geom_point, geom_bar

base = pd.read_csv('teste.csv')

ggplot(base) + aes(x="actor") + geom_bar()
```

```python
import pandas as pd
from plotnine import ggplot, aes, geom_histogram

base = pd.read_csv('teste.csv')

ggplot(base) + aes(x="actor") + geom_histogram(bins=10)
```

```python
import pandas as pd
from plotnine import ggplot, aes, geom_boxplot

base = pd.read_csv('teste.csv')

(
  ggplot(base)
  + aes(x="actor", y="movie")
  + geom_boxplot()
)
```

```python
import pandas as pd
from plotnine import ggplot, aes, geom_bar, coord_flip
base = pd.read_csv('teste.csv')

ggplot(base) + aes(x="actor") + geom_bar() + coord_flip()
```
